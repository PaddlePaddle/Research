# Demo for the data augmentation methods

Read a fundus image and visualize it.

```python
import os
import cv2
import matplotlib.pyplot as plt

image_file = 'demo.jpg'

# Read image file and convert it to RGB order.
image = cv2.imread(image_file)
image = image[:, :, ::-1]

# Visualize the demo image.
plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.axis('off')
plt.title('Original image')
```

Channel-wise random gamma correction.

```python
from data_aug import random_gamma_correction

image_g = random_gamma_correction(image)

# Visualize the demo image.
plt.figure(figsize=(8, 8))
plt.imshow(image_g)
plt.axis('off')
plt.title('With channel-wise random gamma correction')
```

Channel-wise random vessel augmentation.

```python
from data_aug import random_vessel_augmentation

image_v = random_vessel_augmentation(image)

# Visualize the demo image.
plt.figure(figsize=(8, 8))
plt.imshow(image_v)
plt.axis('off')
plt.title('With channel-wise random vessel augmentation')
```
